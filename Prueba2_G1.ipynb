## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes
2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

5. ¿Cuál es el pais con más participación femenina?

```python
#1. Importarlo como nombre athletes
import pandas as pd

athletes = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv")
athletes
```

```python
#2.Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.
athletes = athletes[athletes.gold + athletes.silver + athletes.bronze >= 1].reset_index()
athletes
```

```python
#3.Calcule la masa muscular por fila con una función como: Im.m.=WeightHeight2
athletes.loc[:,"imc"] = athletes.weight / athletes.height**2
athletes
```

```python
#4.Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?
#Se agrupa por deporte y se obtiene la media del peso
max_weight = athletes.groupby("sport")["weight"].mean()

#Con la función idxmax() se obtiene el deporte con maximo de peso y con la función max() el valor máximmo
 print(f"El deporte con mayor peso promedio es {max_weight.idxmax()} con {max_weight.max()}")
```

```python
#5. ¿Cuál es el pais con más participación femenina?
#Se filtra primero las atletas femeninas y luego se agrupa por nacionalidad obteniendo el tamaño-size
athletes_fem = athletes[athletes["sex"] == "female"]
nationality = athletes_fem.groupby("nationality").size()

#Con la función idxmax() se obtiene el pais con maximo de peso y con la función max() el valor máximmo
print(f"El país con el mayor número de atletas femeninas es {nationality.idxmax()} con {nationality.max()} atletas.")
```

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.<
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

R/: 3.1

```python
#importar libreria pandas
import pandas as pd
```

```python
#cargar desde github el dataser nba
df = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
df
```

```python
#eliminar columna unnamed: 0
df.drop(columns = ['Unnamed: 0'], inplace = True)
df
```

```python
#Eliminar datos NaN
df.dropna(inplace = True)
df.reset_index()
df
```

```python
#normalizar nombre columnas en minusculas
df.columns = df.columns.str.lower()
df.head()
```

```python
#agrupar columna team y sumar los salarios de cada equipo en orden ascendente
nba_salary = df.groupby('team')['salary'].agg(['sum']).sort_values(by='sum',ascending=False)
nba_salary
```

R/: 3.2 Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.

```python
#Crear nueva columna, con la descripcion segun la edad de cada jugador
df['Age-Group'] = pd.cut(df.age, bins = 3, labels=['joven','Adulto','Veterano'])
df.sample(15)
```

R/: 3.3

```python
#seleccionar los 5 equipos que mas salario pagan y realizar el index para entregar una lista con los nombres de los equipos
top_teams = df.groupby('team')['salary'].agg(['sum']).nlargest(5, 'sum').index
top_teams
```

```python
#filtar le dataframe mostrando solo los equipos indexados(los 5 equipos que pagan mas salario)
df_top_teams = df[df['team'].isin(top_teams)]
df_top_teams
```

```python
#mostrar la tabla dinamica con los equipos que mas salario pagan, mostrando por posicion de jugador
dinamic_table = df_top_teams.pivot_table(index='team', columns='position', values='salary', aggfunc='mean')
dinamic_table
```

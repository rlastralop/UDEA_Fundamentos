```python
def factorial(n):
    """
    Calcula el factorial de un número entero no negativo.

    Parameters:
    - n (int): Número entero no negativo.

    Returns:
    - int: Factorial de n.
    """
    return np.prod(np.arange(1, n + 1))

def calcular_e(tol=1e-6):
    """
    Calcula una aproximación del número e mediante la serie infinita:
    e = 1/0! + 1/1! + 1/2! + 1/3! + ...

    Parameters:
    - tol (float): Tolerancia para la convergencia de la serie. La función
                  terminará cuando el término actual sea menor que tol.

    Returns:
    - float: Aproximación del número e.
    """
    e_aproximado = 0
    n = 0
    termino = 1.0

    while np.abs(termino) > tol:
        e_aproximado += termino
        n += 1
        fact = factorial(n)
        termino = 1.0 / fact

    return e_aproximado

e_aproximado = calcular_e()
print("Valor aproximado de e:", e_aproximado)
```

```python
#3. ENcuentre los auto valores de la matriz

#Creación de función para el cálculo de autovalores de una matriz nxn
def calcular_autovalores(matriz):
    """
    Calcula los autovalores de una matriz cuadrada.

    Parámetros:
    matriz: Matriz cuadrada de tamaño nxn.

    Retorna:
    autovalores: Array con los autovalores.

    Uso:
    resultado = calcular_autovalores(matriz)
    """
    autovalores = np.linalg.eigvals(matriz)
    return autovalores

#Creación de la matriz
matriz = np.array([[2, 1, 1], [4,3 ,2], [1, 1, 2]])

#Uso de la función para calcular los autovalores , para mayor información  usar help(calcular_autovalores)
autovalores = calcular_autovalores(matriz)

#Visualización de resultados
print("Autovalores de la matriz:", autovalores)
```
